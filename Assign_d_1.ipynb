Ques 1: scraping from webpage

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
```

```python
url = 'https://en.wikipedia.org/wiki/Mount_Everest'
page= urlopen(url)
html= page.read().decode("utf-8")
s_1= BeautifulSoup(html, "html.parser")
s_2=s_1.find('div',class_="mw-content-ltr mw-parser-output")
s_3=s_2.find_all('p')
# soup.title.string
print(s_3)
```

```python
print(type(s_3))
```

```python
s_3=str(s_3)
final=re.sub('<.*?>',' ',s_3)
print(final)
```

```python
s_1=str(s_1)
start_index = s_1.find('<title>')
end_index = s_1.find('</title>')
#storing title in a variable
title=s_1[start_index+7:end_index]
```

```python
#find total count of 'Mount'in the text
final.count('Mount')
```

```python
final.count('mount',re.IGNORECASE)
```

```python
#save text in a file in uppercase
final_1=final.upper()
final_1
```

```python
final_2=re.sub('\n',' ',final_1)
final_2
```

```python
file= open("mount.txt",'w', encoding='utf-8')
file.write(title)
file.write(final_2)
#the output gives the number of characters written to the file.
```

```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read the dataset
with open('/content/sample_data/sarcasm-clusters.txt', 'r') as file:
    content = file.read()

content = list(content.split('\n'))
content
```

```python
df = pd.DataFrame(content, columns=['tweets'])
df
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer

nltk.download('stopwords')
```

```python
swords = stopwords.words('english')
ls = LancasterStemmer()
```

```python
#define a function for text cleaning
def clean_text(sent):
  t1= word_tokenize(sent)
  t2= [x.lower() for x in t1 if x.isalpha()]       #lowercasing
  t3= [x for x in t2 if x not in swords]          #stopword removal
  t4= [ls.stem(x) for x in t3]                    #stemming
  return t4
```

```python
cleaned = df['tweets'].apply(lambda a:clean_text(a))
```

vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(analyzer = clean_text)
```

```python
x_vect = tfidf.fit_transform(df['tweets'])
```

```python
x_vect.shape
```

```python
tfidf.get_feature_names_out()
```

build clustering model

```python
#import the class
from sklearn.cluster import KMeans
```

```python
#create the object
km = KMeans(random_state=0, n_clusters=2)
```

```python
#predict the label for news
labels = km.fit_predict(x_vect)
```

extract most frequent words

```python
from nltk import FreqDist
```

```python
#for label 0
zero = df[labels==0]
```

```python
text = []
for t in zero['tweets']:
  text.extend(clean_text(t))              #words in cluster 2
text
```

```python
freq = FreqDist(text)
```

```python
freq.most_common(10)
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
```

```python
url="https://en.wikipedia.org/wiki/Sri_Lanka"
web=urlopen(url)
tag=web.read().decode("utf-8")
tagged=BeautifulSoup(tag, "html.parser")
tag_1=tagged.find('div',class_='mw-content-ltr mw-parser-output')
tag_2=tag_1.find_all('p')
print(tag_2)     
```

```python
print(type(tag_2))
```

```python
# import nltk
# nltk.download('punkt')
from nltk.tokenize import word_tokenize
nltk.download('punkt')
import pandas as pd
```

```python
sent_1="The postman delivered the package to the wrong address."
sent_2="I wrapped a beautiful present for my friend's birthday."
sent_3="The delivery truck arrived late due to heavy traffic."
sent_4="We need to check the shipping address before sending the order."
sent_5="Online shopping offers a wide variety of products with fast delivery."
```

```python
words = word_tokenize(sent_1) + word_tokenize(sent_2) + word_tokenize(sent_3)+word_tokenize(sent_4)+word_tokenize(sent_5)
save=set(words)
```

```python
#creating empty dataframe
df= pd.DataFrame(columns=list(save),index=[1,2,3,4,5])
df
```

```python
tokens1=word_tokenize(sent_1.lower())
tokens2=word_tokenize(sent_2.lower())
tokens3=word_tokenize(sent_3.lower())
tokens4=word_tokenize(sent_4.lower())
tokens5=word_tokenize(sent_5.lower())
```

```python
tokens1,tokens2,tokens3,tokens4,tokens5
```

```python
count1=[tokens1.count(x) for x in df.columns]
count2=[tokens2.count(x) for x in df.columns]
count3=[tokens3.count(x) for x in df.columns]
count4=[tokens4.count(x) for x in df.columns]
count5=[tokens5.count(x) for x in df.columns]
```

```python
count1,count2,count3,count4,count5
```

```python
df.iloc[0,:]=count1
df.iloc[1,:]=count2
df.iloc[2,:]=count3
df.iloc[3,:]=count4
df.iloc[4,:]=count5
```

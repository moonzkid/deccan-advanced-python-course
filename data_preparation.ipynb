```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
nltk.download('punkt')
import pandas as pd

```

```python
#applicable for all languages
sent_1= 'it was a good practise for us.'
sent_2= 'it was also good to know about it.'
sent_3= 'for all of us they have done a good practise.'
```

```python
#find the list of all distinct words present in these three sentences
# sentence= set(sent_1.split())                       it was a good practise for us
# sentence= set(sent_2.split())                       it was also
# sentence= set(sent_3.split())
# print(sentence)


words=sent_1.split()+sent_2.split()+sent_3.split()
set(words)
```

```python
words = word_tokenize(sent_1) + word_tokenize(sent_2) + word_tokenize(sent_3)
save=set(words)
save
```

```python
#creating empty dataframe
df= pd.DataFrame(columns=list(save),index=[1,2,3])
df
```

```python
tokens1=word_tokenize(sent_1.lower())
tokens2=word_tokenize(sent_2.lower())
tokens3=word_tokenize(sent_3.lower())
```

```python
tokens1,tokens2,tokens3
```

```python
count1=[tokens1.count(x) for x in df.columns]
count2=[tokens2.count(x) for x in df.columns]
count3=[tokens3.count(x) for x in df.columns]
count1,count2,count3
```

```python
df.iloc[0,:]=count1
df.iloc[1,:]=count2
df.iloc[2,:]=count3
df
```

```python
#import count vectorizer class
from sklearn.feature_extraction.text import CountVectorizer
#create the obejct
cv=CountVectorizer()
#apply BOW method
mydata=cv.fit_transform([sent_1,sent_2,sent_3])
mydata.toarray()
```

```python
cv.get_feature_names_out()
```

```python
df=pd.DataFrame(data=mydata.toarray(),columns=cv.get_feature_names_out())
df                    #vectorized counts are visible here
```
